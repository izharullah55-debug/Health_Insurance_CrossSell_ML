**Target Variable Distribution (Graph 1)**

```python
# Target Column = Response
# intrested = 1 and not intrested = 0
plt.figure(figsize=(8,6))
sns.countplot(x='Response',data=df)
plt.title('Customer Response Distribution')
plt.xlabel('Response')
plt.ylabel('Count')
plt.show()
```

**Age Distribution**

```python
plt.figure(figsize=(7,4))
sns.histplot(df['Age'] , bins= 30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

**Gender Respons**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', hue='Response', data=df)
plt.title('Gender Vs Insurance Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Response')
plt.show()
```

**vehicle age vs response**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Vehicle_Age', hue='Response', data=df)
plt.title('Vehicle Age Vs Insurance Distribution')
plt.xlabel('Vehicle Age')
plt.ylabel('Count')
plt.legend(title='Response')
plt.show()
```

**Correlation Heatmap**

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```
